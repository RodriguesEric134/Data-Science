```python
# prompt: criar anlise exploratria da base de dados iris

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Importa a base de dados iris
iris = sns.load_dataset('iris')

# Sumrio estatstico
print(iris.describe())

# Histograma de cada varivel
iris.hist(figsize=(10, 8))
plt.show()

# Boxplot de cada varivel por espcie
for column in iris.columns[:-1]:
    sns.boxplot(x='species', y=column, data=iris)
    plt.title(f'Boxplot de {column} por espcie')
    plt.show()

# Matriz de disperso
sns.pairplot(iris, hue='species')
plt.show()

# Mapa de calor da correlao entre as variveis
sns.heatmap(iris.corr(), annot=True, cmap='coolwarm')
plt.title('Mapa de Calor da Correlao')
plt.show()
```

```python
# prompt: gerar uma anlise exploratria com grficos da base iris

# Importa as bibliotecas necessrias
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Carrega a base de dados iris
iris = sns.load_dataset('iris')

# Sumrio estatstico
print(iris.describe())

# Histograma de cada varivel
iris.hist(figsize=(10, 8))
plt.suptitle('Histogramas das Variveis', fontsize=16)
plt.show()

# Boxplot de cada varivel por espcie
for column in iris.columns[:-1]:
    sns.boxplot(x='species', y=column, data=iris)
    plt.title(f'Boxplot de {column} por Espcie')
    plt.show()

# Matriz de disperso
sns.pairplot(iris, hue='species')
plt.suptitle('Matriz de Disperso', fontsize=16)
plt.show()

# Mapa de calor da correlao entre as variveis
sns.heatmap(iris.corr(), annot=True, cmap='coolwarm')
plt.title('Mapa de Calor da Correlao')
plt.show()

# Violin plot para visualizar a distribuio de cada varivel por espcie
for column in iris.columns[:-1]:
    sns.violinplot(x='species', y=column, data=iris)
    plt.title(f'Violin Plot de {column} por Espcie')
    plt.show()
```
