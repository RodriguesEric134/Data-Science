```python
# prompt: criar análise exploratória da base de dados iris

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Importa a base de dados iris
iris = sns.load_dataset('iris')

# Sumário estatístico
print(iris.describe())

# Histograma de cada variável
iris.hist(figsize=(10, 8))
plt.show()

# Boxplot de cada variável por espécie
for column in iris.columns[:-1]:
    sns.boxplot(x='species', y=column, data=iris)
    plt.title(f'Boxplot de {column} por espécie')
    plt.show()

# Matriz de dispersão
sns.pairplot(iris, hue='species')
plt.show()

# Mapa de calor da correlação entre as variáveis
sns.heatmap(iris.corr(), annot=True, cmap='coolwarm')
plt.title('Mapa de Calor da Correlação')
plt.show()
```

```python
# prompt: gerar uma análise exploratória com gráficos da base iris

# Importa as bibliotecas necessárias
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Carrega a base de dados iris
iris = sns.load_dataset('iris')

# Sumário estatístico
print(iris.describe())

# Histograma de cada variável
iris.hist(figsize=(10, 8))
plt.suptitle('Histogramas das Variáveis', fontsize=16)
plt.show()

# Boxplot de cada variável por espécie
for column in iris.columns[:-1]:
    sns.boxplot(x='species', y=column, data=iris)
    plt.title(f'Boxplot de {column} por Espécie')
    plt.show()

# Matriz de dispersão
sns.pairplot(iris, hue='species')
plt.suptitle('Matriz de Dispersão', fontsize=16)
plt.show()

# Mapa de calor da correlação entre as variáveis
sns.heatmap(iris.corr(), annot=True, cmap='coolwarm')
plt.title('Mapa de Calor da Correlação')
plt.show()

# Violin plot para visualizar a distribuição de cada variável por espécie
for column in iris.columns[:-1]:
    sns.violinplot(x='species', y=column, data=iris)
    plt.title(f'Violin Plot de {column} por Espécie')
    plt.show()
```
