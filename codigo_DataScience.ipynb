```python
#impotando as bibliotecas qeu serão utilizadas
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Cities1.csv')
# Contar o número de valores inválidos das colunas
nan_counts = df.isna().sum()
# Exibe o resultado
print(nan_counts)
```

# Estatiscas Descritivas

```python
# Calculando a moda para a coluna 'Country'
moda_Country = df['Country'].mode()[0]  # Calcula a moda (o valor mais frequente)

# Calculando as frequências absolutas e relativas
freq_absolutas = df['Country'].value_counts()  # Contagem de cada categoria

# Exibindo os resultados
print("Métricas Básicas para a Coluna 'Country':")
print(f"Moda: {moda_Country}")
print("\nFrequências Absolutas:\n", freq_absolutas)
print(f"\n Descrição: {freq_absolutas.describe()}")
```

```python
# Encontrando o valor mais comum na coluna 'Region'
moda_Region = df['Region'].mode()[0]  # Obtém a moda (valor mais recorrente)

# Calculando as contagens absolutas e percentuais
freq_absolutas = df['Region'].value_counts()  # Contagem de cada valor único

# Mostrando os resultados
print("Estatísticas Básicas para a Coluna 'Region':")
print(f"Moda: {moda_Region}")
print("\nFrequências Absolutas:\n", freq_absolutas)
print(f"\nResumo Estatístico:\n {freq_absolutas.describe()}")
```

```python
# Encontrando o valor mais comum na coluna 'City'
moda_cidade = df['City'].mode()[0]  # Obtém a moda (valor mais recorrente)

# Calculando as contagens absolutas e percentuais
freq_absolutas = df['City'].value_counts()  # Contagem de cada valor único

# Mostrando os resultados
print("Análise Estatística para a Coluna 'City':")
print(f"Moda: {moda_cidade}")
print("\nFrequências Absolutas:\n", freq_absolutas)
print(f"\nResumo Estatístico:\n {freq_absolutas.describe()}")
```

```python
# Calculando as estatísticas descritivas para a coluna 'AirQuality'
AirQuality_desc = df['AirQuality'].describe()
print(AirQuality_desc)

media_AirQuality = df['AirQuality'].mean()  # Obtém a média
mediana_AirQuality = df['AirQuality'].median()  # Obtém a mediana
std_AirQuality = df['AirQuality'].std()  # Obtém o desvio padrão
moda_AirQuality = df['AirQuality'].mode()[0]  # Obtém a moda (valor mais recorrente)
quartis_AirQuality = df['AirQuality'].quantile([0.25, 0.5, 0.75])  # Obtém os quartis

# Mostrando os resultados
print("\nEstatísticas Básicas para a Coluna 'AirQuality':")
print(f"Média: {media_AirQuality}")
print(f"Mediana: {mediana_AirQuality}")
print(f"Desvio Padrão: {std_AirQuality}")
print(f"Moda: {moda_AirQuality}")
print(f"Quartis: \n{quartis_AirQuality}")
```

```python
# Calculando as estatísticas descritivas para a coluna 'WaterPollution'
WaterPollution_desc = df['WaterPollution'].describe()
print(WaterPollution_desc)

media_WaterPollution = df['WaterPollution'].mean()  # Obtém a média
mediana_WaterPollution = df['WaterPollution'].median()  # Obtém a mediana
std_WaterPollution = df['WaterPollution'].std()  # Obtém o desvio padrão
moda_WaterPollution = df['WaterPollution'].mode()[0]  # Obtém a moda (valor mais recorrente)
quartis_WaterPollution = df['WaterPollution'].quantile([0.25, 0.5, 0.75])  # Obtém os quartis

# Mostrando os resultados
print("\nEstatísticas Básicas para a Coluna 'WaterPollution':")
print(f"Média: {media_WaterPollution}")
print(f"Mediana: {mediana_WaterPollution}")
print(f"Desvio Padrão: {std_WaterPollution}")
print(f"Moda: {moda_WaterPollution}")
print(f"Quartis: \n{quartis_WaterPollution}")
```

# Graficos das Medias da Poluição do Ar



```python
df = pd.read_csv('Cities1.csv')
df = df[df['Country'] =='Brazil']
df_reg_air = df[['Region', 'AirQuality']]

df_br = df_reg_air.groupby('Region').mean()
df_br
```

```python
# Visualização de Dados
## Criando gráficos para insights
### Exemplo de Gráfico de Barras
plt.figure(figsize=(80, 20))
df_br.plot(kind='bar')
plt.title('Média da Qualidade do Ar por Estado Brasileiro')
plt.xlabel('Estado')
plt.ylabel('Média da Qualidade do Ar')
plt.show()
```
