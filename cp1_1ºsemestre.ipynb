```python
import pandas as pd
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.model_selection import train_test_split
```

```python
base = pd.read_csv('onlinefoods.csv')
```

```python
base
```

```python
base  = base.drop(['Unnamed: 12'], axis = 1)
base
```

```python
base_categoricos  = base[['Gender','Marital Status','Occupation','Monthly Income','Educational Qualifications','Output']]
base_categoricos = pd.get_dummies(base_categoricos, dtype=int)
```

```python
base_numericos  = base.drop(['Gender','Marital Status','Occupation','Monthly Income','Educational Qualifications','Output','Feedback'], axis = 1)
```

```python
base_modelo = pd.concat([base_categoricos, base_numericos],axis = 1)
```

```python
y = base['Feedback']
```

```python
X_train, X_test, y_train, y_test = train_test_split(base_modelo, y, train_size = 0.7)
```

```python
classificador = DecisionTreeClassifier()
```

```python
classificador.fit(X_train, y_train)
```

```python
classificador.score(X_test,y_test)
```
